```python
# main.ipynb
# YOLO 模型推論主程式

# 1. 安裝並匯入必要套件
!pip install ultralytics opencv-python matplotlib --quiet

import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import os

# 2. 確認工作目錄
print("Current working directory:", os.getcwd())
```

```python
# 載入你訓練好的模型 (weights/best.pt)
model = YOLO("weights/best.pt")
print("模型載入完成！")
```

```python
# 測試推論
test_image_path = "asset/train_batch0.jpg"  # 這裡換成你的測試圖片
results = model.predict(source=test_image_path, save=False)  # 不存檔，只在記憶體顯示

# 顯示結果
result_img = results[0].plot()  # YOLO 內建的結果繪圖
plt.imshow(cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
output_path = "asset/output_result.jpg"
cv2.imwrite(output_path, result_img)
print(f"結果已儲存至: {output_path}")
```
